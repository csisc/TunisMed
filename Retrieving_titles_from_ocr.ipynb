```python
#Getting text size and page for every PDF page
#Source code: Tayyab Khan Satti, https://medium.com/@i190712/how-to-extract-text-from-pdf-on-the-basis-of-font-size-using-python-libraries-e931e583749b
from pdfminer.high_level import extract_pages, extract_text
from pdfminer.layout import LTTextContainer, LTChar,LTLine,LAParams
import os

year = "2008"
path= "/content/drive/MyDrive/" + year + "_1.pdf"

Extract_Data=[]
pg = 0
for page_layout in extract_pages(path):
    pg += 1
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            b = False
            for text_line in element:
              try:
                s = [x for x in text_line]
              except:
                b = True
            if (b == False):
              for character in text_line:
                try:
                  Font_size=character.size
                except:
                  print("character")
            Extract_Data.append([pg, Font_size,(element.get_text())])
```

```python
#Retrieving titles from Extract_Data
titles = []
for i in range(len(Extract_Data)-1):
  if (Extract_Data[i+1][1] - Extract_Data[i][1] > 1.5):
    if (Extract_Data[i][1] == Extract_Data[i-1][1]):
      titles.append(Extract_Data[i-1][0], Extract_Data[i-1][2])
    titles.append(Extract_Data[i][0], Extract_Data[i][2])
```
